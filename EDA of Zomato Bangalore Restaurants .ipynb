## Data Manipulation

```python
# Droping all unwanted columns such as 'url','address','phone','dish_liked','reviews_list','menu_item'.

df = df.drop(['url','address','phone','dish_liked','reviews_list','menu_item'], axis = 1)

df.head()
```

```python
df.info()
```

```python
# Dropping Duplicates

df.drop_duplicates(inplace = True)
df.shape
```

```python
df.head()
```

```python
# Checking unique values in Rate column

df['rate'].unique()
```

```python
# cleaning Rate column

def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
# finding Null values

df['rate'].isnull().sum()
```

#### note : There are total '10019' null values found, we simply cannot drop this much number of amount. So insted of droping the amount we going to replace it with 'Mean'.


```python
# replacing null with 'mean'

df['rate'].fillna(df['rate'].mean(), inplace = True)
df['rate'].isnull().sum()
```

```python
df.info()
```

```python
# Droping rest null values

df.dropna(inplace = True)
df.head()
```

```python
df.info()
```

#### Note : as we can check there is no null value in dataset now.

```python
# renaming the 'approx_cost(for two people)' as 'cost2plates' and 'listed_in(type)' as 'Type'.

df.rename(columns = {'approx_cost(for two people)' : 'Cost2plates','listed_in(type)':'Type'}, inplace = True)

df.head()
```

```python
# checking for unique value in location 

df['location'].unique()
```

```python
# checking for unique value in location 

df['listed_in(city)'].unique()
```

#### Note : ''listed_in(city)' and 'location' both are same so lets keep only one of them and drop the other.

```python
# Dropping the 'listed_in(city)' column.

df = df.drop('listed_in(city)', axis = 1)
df.head()
```

```python
# Finding unique values in 'Cost2plates'

df['Cost2plates'].unique()
```

```python
# Removing comma from the values of 'Cost2plates'

def handlecomma(value):
    value= str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)

df['Cost2plates'] = df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()
```

#### Note : now we can check there is no comma in middle of value.

```python
# cleaning 'rest_type' column

rest_types = df['rest_type'].value_counts(ascending = False)

rest_types
```

```python
# creating dataframe for restuarent types of which frequency is less than 1500

rest_types_less_than1500 = rest_types[rest_types<1500]
rest_types_less_than1500
```

```python
# Creating rest_type which are less than 1500 as 'others'.

def handle_rest_type(value):
    if(value in rest_types_less_than1500):
        return 'others'
    else:
        return value
    
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
# creating dataframe for 'location' of which frequency is less than '500'

location = df['location'].value_counts()
location_less_than_500 = location[location<500]
location_less_than_500
```

```python
# Creating rest_type which are less than 500 as 'others'.

def handle_location(value):
    if(value in location_less_than_500):
        return 'others'
    else:
        return value
    
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
# creating dataframe for 'cuisines' of which frequency is less than '100'

cuisines = df['cuisines'].value_counts()

cuisines_less_than_100 = cuisines[cuisines<100]
```

```python
# adding 'cuisines_less_than_100' as others

def handle_cuisines(value):
    if(value in cuisines_less_than_100):
        return 'others'
    else:
        return value
    
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
